```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from datetime import datetime as dt 
import os

location = '/Users/ypi/Desktop/Hands-On-Explainable-AI-XAI-with-Python-Chapter01/Location_History.json'

read_location = pd.read_json(location)
print('There are %d rows in the location history' %len(read_location))
```

```python
read_location['lat'] = read_location['locations'].map(lambda x:x['latitudeE7'])
read_location['lon'] = read_location['locations'].map(lambda x:x['longitudeE7'])
read_location['timestamp_ms'] = read_location['locations'].map(lambda x:x['timestampMs'])
```

```python
read_location['lat'] = read_location['lat'] / 10.**7
read_location['lon'] = read_location['lon'] / 10.**7

read_location['timestamp_ms'] = read_location['timestamp_ms'].astype(float) / 1000
read_location['datetime'] = read_location['timestamp_ms'].map(


lambda x: dt.fromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))

date_range = '{}-{}'.format(read_location['datetime'].min()[:4], read_location['datetime'].max()[:4] )
```

```python
read_location = read_location.drop(labels=['locations', 'timestamp_ms'], axis=1, inplace=False)
```

```python
import pandas as pd
import folium
from folium import plugins
# San Francisco latitude and longitude values
latitude = 48.8
longitude = 2.3
# create map and display it
paris_map = folium.Map(location=[latitude, longitude], zoom_start=12)


latitudes = list(read_location['lat'])
longitudes = list(read_location['lon'])



for lat, lng in zip(latitudes[:1000], longitudes[:1000]):
    folium.Marker([lat, lng]).add_to(paris_map) 
 


# display the map of San Francisco
paris_map
```
