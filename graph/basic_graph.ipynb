```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = np.random.randn(5,3)
df = pd.DataFrame(np.abs(data), 
                  index=['Mon', 'Tue', 'Wen', 'Thir', 'Fri'], 
                  columns=['A', 'B', 'C'])
df
```

```python
df.plot()
plt.show()
## dataframe.plot  
## the line shows the column, and the x-axis is the row
```

```python
df.plot(kind='barh')
```

```python
df.plot(kind='barh',stacked=True)
## set stacked as True
```

```python
# x: x
# y: y
# s: 
# c: 
# : help(df.plot.scatter)
df.plot(kind='scatter', x='A', y='B', s=df.C*500, c='r')
plt.show()
```

```python
df.plot(kind='box')
```

```python
df['A'].plot(kind='bar')
```
