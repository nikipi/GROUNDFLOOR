```python
from bs4 import BeautifulSoup
from mechanize import Browser

#This retrieves the webpage content
br = Browser()
br.set_handle_robots(False)

import ssl
ssl._create_default_https_context = ssl._create_unverified_context


############# you can change the res here to the target url 
res = br.open("https://www.ecb.europa.eu/press/key/speaker/bm/html/index.en.html")
data = res.get_data() 

#This parses the content
soup = BeautifulSoup(data)
titles = soup.find_all('div',{'class':"title"})

#This outputs the content :)
weblinks=[]
speechtitle=[]
for title in titles:
    if title.find('a')!=None:
        weblinks.append('https://www.ecb.europa.eu'+title.find('a')['href'])
        speechtitle.append(title.find('a').contents[0])

        
```

```python
# build a dataframe

import pandas as pd
df=pd.DataFrame({'title':speechtitle,'links':weblinks},columns=['title','links'])
#only keep html link
df=df[~df.links.str.contains("slides")]
df=df[~df.links.str.contains("pdf")]
# get year info from the link and have a new column
df['year']=df.links.str.split('/').str[6]
df['speaker']=df.title.str.split(':').str[0]
df=df.loc[df.year>'2017']
```

```python
pd.options.display.max_rows = 999
df
```

```python
text=[]
for row in df.links:
    
    url=row
    soup= get_soup_html(url)
    a=[link.string for link in soup.find_all('p')]
    text.append(a[:-6])
      
```

```python
# add the content to the dataframe
content = pd.Series(text)
df['text'] = content.values
df.to_csv('/Users/ypi/Desktop/ecb_pred.csv', index = False, header=True)
```
