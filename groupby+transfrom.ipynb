```python
import numpy as np
import pandas as pd
df=pd.DataFrame({'key':['a','b','c']*4,
                'value':np.arange(12.)})
```

```python
df
```

```python
g=df.groupby('key')
```

```python
g.mean()
```

```python
g=df.groupby('key').value
```

```python
g.mean()
```

```python
g.transform('mean')
```

```python
def normalize(x):
    return (x-x.mean()/x.std())
```

```python
g.transform(normalize)
```

```python
g.apply(normalize)
```
