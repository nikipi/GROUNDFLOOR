```python
from parse import parse
```

```python
eg1=parse('if you are {},you can have {}','if you are 21 years old,you can have wine')
print(eg1)
```

```python
eg1[0]
```

```python
eg1[1]
```

```python
eg2=parse('if you are {age},you can have {drinks}','if you are 21 years old,you can have wine')
```

```python
print(eg2)
```

```python
type(eg2['age'])
```

```python
eg4=parse('if you are {age:d} years old,you can have {drinks}','if you are 21 years old,you can have wine')
```

```python
type(eg4['age'])
```

### resue the pattern

```python
from parse import compile
```

```python
pattern=compile('my name is {},{} years old,{}')
```

```python
pattern.parse('my name is jack,17 years old,male')
```

```python
pattern.parse('my name is niki,21 years old,female')
```

## delete the space

```python
parse('hello {:^} , hello python', 'hello     world    , hello python')
```

```python
parse('hello {:>} , hello python', 'hello     world    , hello python')
```

```python
parse('hello {:<} , hello python', 'hello     world    , hello python')
```

```python
parse("{}@gmail",'2345@gmail')
```

```python
parse("{}@gmail",'2345@Gmail',case_sensitive=True)
```
