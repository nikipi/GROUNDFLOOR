```python
import difflib
hd = difflib.HtmlDiff()
loads = ''
with open('/Users/apple/Desktop/1.csv','r') as load:
    loads = load.readlines()
    load.close()

mems = ''
with open('/Users/apple/Desktop/2.csv', 'r') as mem:
    mems = mem.readlines()
    mem.close()

with open(' htmlout.html','a+') as fo:
    fo.write(hd.make_file(loads,mems))
    fo.close()
```
